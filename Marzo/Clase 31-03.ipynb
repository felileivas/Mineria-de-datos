```python
#import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

def tofahrenheit(dato):
    return 9.0/5.0 * dato +32

celsius = pd.DataFrame(np.arange(15,28))
farenheit = pd.DataFrame(tofahrenheit(celsius))

result = pd.concat([celsius, farenheit], axis=1, join='outer')
df = pd.DataFrame(result)
df.columns = ['Celsius', 'Farenheit']

sns.regplot(x="Celsius", y="Farenheit", data=df);


#sns.jointplot(x="Celsius", y="Farenheit", data=df, kind="reg");
#lmplot('Celsius', 'Farenheit', data=df)


#plot(celsius, farenheit)
#plt.grid()
#plt.xlabel("Grados farenheit")
#plt.ylabel("Grados Celsius")
#plt.title('Grafico de celsius a farenheit')


```
