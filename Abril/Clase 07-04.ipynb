```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

datos = pd.read_csv("netflix_titles.csv")
df=pd.DataFrame(datos)
df
```

```python
df.describe()
```

```python
len(df.show_id.unique())
```

```python
df_movie = df[df.type == 'Movie']
df_tvshow = df[df.type == 'TV Show']
df_movie.count()
```

```python
dfnew =pd.crosstab(df.release_year, df.type)
dfnew
sns.lmplot(x='Movie', y='TV Show', data=dfnew)
#sns.regplot(x="Movie", y="TV Show", data=dfnew);
#sns.lmplot(x=df.Attack, y=df.Defense)
```

```python
sns.regplot(x="Movie", y="TV Show", data=dfnew);
```

```python
df_datos = df[df.release_year > 2010]
df_agrupado = df_datos.groupby(['release_year','type'])[['type']].count()
```

```python
sns.countplot(x='release_year', hue='type', data=df_datos)
```

```python
sns.catplot(x="release_year",col="type",palette="Set1", kind="count", data=df_datos)
```

```python
df_datos.type.value_counts().plot(kind='pie')
plt.title('Tipo de produccion audivisual por tipo')
```

```python
for t_type in ['Movie','TV Show']:
    df_datos.release_year[df_datos.type == t_type].plot(kind="kde")
    
plt.legend({"Movie", "TV Show"})
plt.title('Distribucion por ao')
```
