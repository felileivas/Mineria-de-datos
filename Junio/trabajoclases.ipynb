```python
import collections

arreglo_tipos = []
cuenta = collections.Counter(train.Item_Type.tolist())

for itemTipo in cuenta:
    arreglo_tipos.append(itemTipo)
    
arreglo_numero = []
    
for numeroTipo in range(1,16+1):
    arreglo_numero.append(numeroTipo)
    
arreglo_tipos
arreglo_numero

 
train['Item_Type'].replace(arreglo_tipos,arreglo_numero,inplace=True)

train.groupby(['Item_Type']).agg({'Item_Identifier':'count'})
```

```python
import collections

def arrayListItems(cuenta):
    arreglo_tipos = []  
    for itemTipo in cuenta:
        arreglo_tipos.append(itemTipo)
    return arreglo_tipos

arreglo_numero = []
    
for numeroTipo in range(1,len(arreglo_tipo)+1):
    arreglo_numero.append(numeroTipo)
    
arreglo_tolist = collections.Counter(train.Outlet_Establishment_Year.tolist())
arreglo_tipo = arrayListItems(arreglo_tolist)

train['Outlet_Establishment_Year'].replace(arreglo_tipo,arreglo_numero,inplace=True)
train.groupby(['Outlet_Establishment_Year']).agg({'Item_Identifier':'count'})
```
