# K Means

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets.samples_generator import make_blobs
```

```python
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
print("X: ",X.shape)
print("y_true: ",y_true.shape)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=20);
```

```python
inertia = list()
for i in range(1, 10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    inertia.append([i, kmeans.inertia_])
inertia
```

```python
# x es el numero de n_cluster (del 1 al 9)
# y es el resultado del modelo
x, y = zip(*[inertia for inertia in inertia])
```

```python
plt.plot(x, y, 'ro-', markersize=8, lw=2)
plt.grid(True)
plt.xlabel('Numero de Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(X)
# prediciendo los grupos
y_kmeans = kmeans.predict(X)
```

```python
y_true
```

```python
y_kmeans
```

```python
centroides = kmeans.cluster_centers_
```

```python
colores=['red','green','blue','black']
asignar=[]
```

```python
for i in y_kmeans:
    asignar.append(colores[i])
```

```python
plt.scatter(X[:, 0], X[:, 1],  c = asignar,s=30)
plt.scatter(centroides[:, 0], centroides[:, 1],  marker = '*', c = ['yellow'], s=200)
plt.show()
```

# Cluster Gerarquico

### DENDOGRAMA

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'),orientation='right')

plt.title('Dendograma')
plt.xlabel('Grupos')
plt.ylabel('Distancias Euclidianas')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters = 4, 
                    affinity = 'euclidean', 
                    linkage = 'ward')

y_ac = ac.fit_predict(X)
```

```python
# Visualising the clusters
plt.scatter(X[y_ac == 0, 0], X[y_ac == 0, 1], s = 30, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_ac == 1, 0], X[y_ac == 1, 1], s = 30, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_ac == 2, 0], X[y_ac == 2, 1], s = 30, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_ac == 3, 0], X[y_ac == 3, 1], s = 30, c = 'cyan', label = 'Cluster 4')
plt.title('Grupos')
plt.legend(loc='best')
plt.show()
```
